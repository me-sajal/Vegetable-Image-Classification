```python
def view_random_image(target_dir, target_class,verbose=0):

    if target_class==None:
        target_class=random.sample(os.listdir(train_data_dir), 1)[0]

    # setting up the image directory
    target_folder = target_dir + "/" + target_class

    #get a random image path
    random_image = random.sample(os.listdir(target_folder), 1)

    #read image and plotting it
    img = mpimg.imread(target_folder + "/" + random_image[0] )
    plt.imshow(img)
    plt.title(target_class)
    plt.axis("off")

    if verbose==1:
        print(f"Image shape: {img.shape}")

    return img
```

# Tensorflow Keras

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, GlobalAveragePooling2D, Rescaling
```

This code snippet imports necessary modules from the TensorFlow library for building and training deep learning models. It imports components for building a Sequential model, various layers such as Conv2D (2D convolutional layer), MaxPooling2D (2D max-pooling layer), Dense (fully connected layer), Dropout (regularization layer), GlobalAveragePooling2D (global average pooling layer), and Rescaling (input data rescaling layer). This code prepares the tools required to define and train a convolutional neural network (CNN) or similar deep learning models.

# Model Creation


```python
# Model Creation
model = Sequential()
# model.add(Rescaling(1./255))
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(GlobalAveragePooling2D())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dense(vegtable_classes, activation='softmax'))
model.summary()

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

```

This code defines and compiles a convolutional neural network (CNN) model for a computer vision task.

1. It creates a sequential neural network model.

2. It adds three convolutional layers with increasing numbers of filters and ReLU activation functions, followed by max-pooling layers to reduce the spatial dimensions.

3. After the convolutional layers, it adds a Global Average Pooling layer to reduce the spatial dimensions further.

4. It adds two dense (fully connected) layers with ReLU activation functions and a dropout layer to prevent overfitting.

5. Finally, it adds an output dense layer with a softmax activation function, which is typically used for multi-class classification tasks. The number of neurons in this layer depends on the variable `vegtable_classes`, representing the number of classes in the classification problem.

6. The model is summarized to display its architecture.

7. The model is compiled with the Adam optimizer, sparse categorical cross-entropy loss (suitable for classification tasks with integer labels), and accuracy as the evaluation metric.

# Model Fit

```python
import tensorflow as tf
callbacks = [
  tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    verbose=2,
    restore_best_weights=True
  )
]
# , tf.keras.callbacks.ModelCheckpoint('model.h5', monitor='loss',save_best_only=True)

history = model.fit(train_data, epochs=20, validation_data=val_data, callbacks=callbacks)
```

This code is using the TensorFlow library to train a neural network model.

1. It imports the TensorFlow library.

2. It sets up a callback called EarlyStopping. This callback monitors the validation loss during training and stops training early if the validation loss doesn't improve for a certain number of epochs (in this case, 5 epochs). It also restores the best weights of the model when training stops.

3. It creates a list called "callbacks" and adds the EarlyStopping callback to it.

4. It trains a neural network model (which should be defined elsewhere in the code) using training data (train_data) for 20 epochs. During training, it also uses validation data (val_data) for monitoring the model's performance.

5. It specifies the "callbacks" list to be used during training, which means that the EarlyStopping callback will be invoked during the training process to potentially stop training early if the validation loss doesn't improve.

```python
# Evaluate the model after fine-tuning
test_loss, test_accuracy = model.evaluate(test_data, steps=len(test_data))
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

3. The predictions are stored in the `predictions` variable.

4. The code extracts the class with the highest predicted probability (argmax) for each image using `np.argmax(predictions, axis=1)` and appends these predicted labels to an existing list called `predicted_labels`.

5. It also extracts the true labels (ground truth) from the `labels` variable and appends them to an existing list called `true_labels`.

In summary, this code evaluates a machine learning model's performance by making predictions on a test dataset and collecting both the predicted labels and true labels for later analysis, such as calculating accuracy or other evaluation metrics.

```python
# Create a classification report
class_names = sorted(os.listdir(train_data_dir))
report = classification_report(true_labels, predicted_labels, target_names=class_names)
print(report)
```

   - `true_labels`: The actual labels or ground truth values for your dataset.
   - `predicted_labels`: The predicted labels generated by your machine learning model.

3. `target_names`: It specifies the names associated with each class in the classification problem. In this code, it's set to the `class_names` obtained earlier.

```python
# Create a confusion matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)

plt.figure(figsize=(10, 8))
plt.imshow(conf_matrix, cmap=plt.cm.Blues)

# Show numbers in the confusion matrix
for i in range(len(class_names)):
    for j in range(len(class_names)):
        plt.text(j, i, str(conf_matrix[i, j]), ha='center', va='center', color='white' if conf_matrix[i, j] > np.max(conf_matrix) / 2 else 'black')

plt.title('Confusion Matrix')
plt.xticks(np.arange(0, len(class_names)), class_names, rotation=45)
plt.yticks(np.arange(0, len(class_names)), class_names)
plt.colorbar()
plt.savefig("confusion_matrix.png")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

This code imports three functions (precision_score, recall_score, and f1_score) from the sklearn.metrics module. These functions are commonly used in machine learning and statistics to calculate different performance metrics for classification models, such as precision, recall, and F1-score. These metrics are used to evaluate how well a model is performing in terms of its ability to correctly classify data points.

```python
precision = precision_score(true_labels, predicted_labels, average ='micro')
recall    = recall_score(true_labels, predicted_labels, average ='micro')
f1        = f1_score(true_labels, predicted_labels, average ='micro')

print(f"Precision = {round(precision * 100, 2)}%")
print(f"Recall = {round(recall * 100, 2)}%")
print(f"F1-score = {round(f1 * 100, 2)}%")
```

This code calculates and prints the precision, recall, and F1-score for a classification task.  

1. `precision_score`, `recall_score`, and `f1_score` are functions typically used in machine learning for evaluating classification models.

2. These functions take two sets of labels as input: `true_labels` (the actual correct labels) and `predicted_labels` (the labels predicted by a classification model).

3. The code calculates the precision, recall, and F1-score for the micro-average, which means it computes these metrics by considering all the instances in the dataset as a single set, rather than computing separate metrics for each class. This is useful when dealing with imbalanced datasets.

4. The calculated metrics are then rounded and printed as percentages to the console, providing an assessment of how well the model is performing in terms of precision, recall, and the F1-score.
